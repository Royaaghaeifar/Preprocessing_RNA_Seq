```python
import os 
import pandas as pd
from glob import glob

from itertools import compress
```

```python
#Define path
path = os.getcwd()+ '/'
folder_path = os.listdir(path)
```

```python
folder_path.remove('.DS_Store')
folder_path.remove('MANIFEST.txt')
folder_path.remove('CMI_MBC.ipynb')
folder_path.remove('.ipynb_checkpoints')
```

```python
len(folder_path)
```

```python
paths =[]

for folder in folder_path:
    folder += '/'
    sub_folders = os.path.join(path,folder)
    paths.append(sub_folders)
```

```python
tsv_files= []

for path in paths:    
    tsv = glob(os.path.join(path, '*.tsv'))
    tsv_files.append(tsv)
```
